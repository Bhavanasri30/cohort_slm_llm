#RAG (Retrieval-Augmented Generation)
 Definition :

RAG is a technique where an LLM first searches external documents and then uses the retrieved information to generate an answer.

Formula
RAG = Retrieval + LLM Generation


#Types of RAG

There are 3 major types.

**1. Naive RAG (Basic RAG)**

Process:

* Question

* Retrieve

* LLM

Features:

* Simple
* Easy to build
* Basic retrieval

Example: Chat with PDF


**2. Advanced RAG**

Adds improvements.

Process:

* Question

* Retrieve Many Chunks

* Re-Rank Best Chunks

* LLM

Features:

* Better chunking
* Metadata filtering
* Re-ranking
* Higher accuracy

Example:  Enterprise Search Systems



**3. Agentic RAG**

Most advanced.

Process:

* Question

* AI Agent Thinks

* Search Multiple Sources
  
* Combine Results

* LLM

Features:

* Planning
* Reasoning
* Multi-step retrieval
* Tool usage

Example:
AI Research Assistant

# Components of RAG

```text
1. Knowledge Source
2. Text Chunking & Preprocessing
3. Embedding Model
4. Vector Database
5. Query Encoder
6. Retriever
7. Augmentation Layer
8. LLM
```



# 1. Knowledge Source


The place where information comes from.

Examples:

```text
PDFs
Websites
Word Documents
Databases
Books
Company Policies
```

### Example

```text
College Placement Rules.pdf
```

This is the raw knowledge that RAG will use.



# 2. Text Chunking & Preprocessing


Large documents are split into smaller pieces (chunks).



* A 100-page PDF cannot be searched efficiently as one huge block.Hence,they are split into chunks.

### Example

Document:

```text
Placement Rules

Students need 7.5 CGPA.
Students should have no active backlogs.
```

After chunking:

```text
Chunk 1:
Students need 7.5 CGPA.

Chunk 2:
Students should have no active backlogs.
```

### Preprocessing

Cleaning the text:

```text
Remove extra spaces
Remove unwanted symbols
Fix formatting
```



# 3. Embedding Model

Converts text into numbers (vectors).

### Example

```text
Students need 7.5 CGPA.
```

becomes:

```text
[0.24, 0.67, 0.91, ...]
```




# 4. Vector Database

### What is it?

Stores all embeddings (vectors).

Examples:

* ChromaDB
* FAISS
* Pinecone
* Weaviate

### Example

```text
Chunk 1 → [0.24, 0.67, 0.91]
Chunk 2 → [0.12, 0.89, 0.31]
```

Stored in the Vector Database.

### Purpose

Fast similarity search.



# 5. Query Encoder



When a user asks a question, the question must also become a vector.

### Example

User asks:

```text
What CGPA is required?
```

Query Encoder converts it to:

```text
[0.26, 0.64, 0.88]
```

Usually, the same embedding model is used.

### Purpose

Convert the user's query into a format that can be compared with document vectors.



# 6. Retriever

Searches the Vector Database for the most similar chunks.

### Example

Query:

```text
What CGPA is required?
```

Retriever finds:

```text
Students need 7.5 CGPA.
```

### Purpose

Fetch relevant information.



# 7. Augmentation Layer


Combines:

```text
User Question
+
Retrieved Chunks
```

into a prompt for the LLM.

### Example

```text
Question:
What CGPA is required?

Context:
Students need 7.5 CGPA.
```

This step "augments" the prompt with external knowledge.



# 8. LLM



The language model that generates the final answer.

Examples:

* OpenAI GPT
* Google Gemini
* Meta Llama

### Input

```text
Question
+
Retrieved Context
```

### Output

```text
The placement eligibility requirement is 7.5 CGPA.
```



# Complete Flow of RAG

```text
Knowledge Source
       ↓
Chunking & Preprocessing
       ↓
Embedding Model
       ↓
Vector Database
       ↓
User Question
       ↓
Query Encoder
       ↓
Retriever
       ↓
Augmentation Layer
       ↓
LLM
       ↓
Final Answer
```
