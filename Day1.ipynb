```python
#if statement
num=int(input())
if(num>0):
  print("positive number")
```

```python
#if-else statement
num=int(input())
if num%2==0:
  print("Even number")
else:
  print("odd number")
```

```python
a=int(input())
b=int(input())
c=int(input())
if (a>b and a>c):
  print("A is largest")
elif (b>a and b>c):
  print("B is largest")
else:
  print("C is largest")
```

```python
# multiplication table
num=int(input())
for i in range(1,11):
  print(num,"×",i,"=",num*i)
```
