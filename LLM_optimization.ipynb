# LLM Optimization & Evaluation

## Introduction

Large Language Models (LLMs) are powerful AI systems that generate human-like text. However, they require significant computational resources, memory, and processing power. Therefore, optimization techniques are used to improve speed, reduce cost, and manage resources efficiently. Evaluation techniques are used to measure how well a model performs on different tasks.



# PART 1: LLM OPTIMIZATION

Optimization focuses on making LLMs faster, cheaper, and more efficient.



# 1. Tokens

## Definition

A token is the smallest unit of text processed by an LLM.

The model does not read complete sentences directly. It first breaks text into tokens.

## Example

Sentence:

I love Artificial Intelligence.

Possible Tokens:

I | love | Artificial | Intelligence | .

## Why Tokens Are Needed

Computers cannot understand words directly. Tokens allow text to be converted into numerical form for processing.

## How It Works

* Text

* Tokens

* Token IDs

* Embeddings

* LLM Processing

## Real-Life Analogy

Imagine a book.

Humans read complete sentences.

An LLM reads small pieces of those sentences, just like reading the book word by word.







# 2. Tokenization

## Definition

Tokenization is the process of converting text into tokens.

## Example

Input:

ChatGPT is amazing.

Tokens:

Chat | GPT | is | amazing | .

## Why Tokenization Is Needed

The model can only work with numbers, not raw text.

Tokenization converts language into a machine-readable format.

## Real-Life Analogy

Think of tokenization as cutting a pizza into slices before eating it.

The pizza is the sentence.

The slices are the tokens.




# 3. Token Burning

## Definition

Token burning refers to the total number of tokens consumed during model usage.

## Formula

Total Tokens = Input Tokens + Output Tokens

## Example

Input:

Explain AI.

Input Tokens: 20

Output:

AI is a technology that learns patterns from data.

Output Tokens: 80

Total Tokens Burned: 100

## Why It Matters

* Determines API cost
* Affects processing speed
* Consumes context window space

## Real-Life Analogy

Think of tokens as fuel.

More fuel consumed means higher operating cost.



Token burning is the consumption of tokens during input and output processing.



# 4. Context Window

## Definition

The context window is the maximum number of tokens an LLM can process at one time.

## What Fits Inside Context Window

* System Prompt
* Chat History
* User Question
* Retrieved Documents
* Model Response

## Example

Context Window = 100K Tokens

Chat History = 20K

Documents = 40K

Question = 2K

Total Used = 62K

Remaining = 38K

## Why Context Window Is Important

A larger context window allows the model to handle:

* Long conversations
* Large documents
* Research papers
* Complex RAG systems

## Real-Life Analogy

Imagine a whiteboard.

A small whiteboard holds little information.

A large whiteboard holds much more information.

The context window is the size of the whiteboard.



The context window is the maximum amount of information an LLM can process in a single interaction.


# 5. Headroom

## Definition

Headroom is the remaining space in the context window available for generating a response.

## Formula

Headroom = Context Window − Current Input

## Example

Context Window = 100K

Input = 80K

Headroom = 20K

The model can use approximately 20K tokens for output.

## Why Headroom Matters

Without enough headroom:

* Responses become shorter
* Responses may be cut off
* Generation quality decreases

## Real-Life Analogy

Imagine writing an exam.

If your answer sheet is almost full, you have little space left to write.

That remaining space is headroom.



Headroom is the unused portion of the context window reserved for output generation.



# 6. Temperature

## Definition

Temperature controls the randomness and creativity of the model's responses.

## Low Temperature (0–0.3)

Produces:

* Predictable answers
* Consistent responses
* Less creativity

Example:

Question:

What is the capital of France?

Answer:

Paris

Almost every time.

## High Temperature (0.8–1.5)

Produces:

* More creative responses
* More variation
* Higher randomness

Useful for:

* Story writing
* Brainstorming
* Creative content

## Real-Life Analogy

Low Temperature:

Following a recipe exactly.

High Temperature:

Experimenting while cooking.



Temperature controls the randomness and creativity of generated text.



# 7. Top-K Sampling

## Definition

Top-K sampling limits the model to choosing from the K most probable next words.

## Example

Predicted Words:

Paris = 80%

London = 10%

Berlin = 5%

Apple = 1%

If K = 3

The model only considers:

Paris

London

Berlin

## Why It Is Used

Prevents unlikely and nonsensical outputs.

## Real-Life Analogy

Imagine selecting the top 3 candidates from 100 applicants before making a final decision.



Top-K sampling restricts word selection to the K most likely next tokens.



# 8. Top-P Sampling

## Definition

Top-P sampling selects words whose combined probability reaches a chosen threshold P.

## Example

Paris = 80%

London = 10%

Berlin = 5%

Rome = 3%

Top-P = 90%

Selected:

Paris

London

Because:

80% + 10% = 90%

## Why It Is Used

Provides more flexibility than Top-K.

## Real-Life Analogy

Keep selecting candidates until you cover 90% of the required skills.



Top-P sampling selects the smallest set of words whose cumulative probability reaches a specified threshold.



# 9. Latency

## Definition

Latency is the time taken by a model to start responding after receiving a request.

## Example

User asks a question.

The model starts responding after 2 seconds.

Latency = 2 seconds.

## Why It Matters

Lower latency improves user experience.

## Real-Life Analogy

Ordering food at a restaurant.

The waiting time before receiving food is latency.



Latency is the delay between a user request and the beginning of the model's response.



# 10. Throughput

## Definition

Throughput measures how much work a model can handle within a given period.

## Example

A model processes:

100 requests per second.

Throughput = 100 requests/second.

## Why It Matters

Higher throughput allows more users to use the system simultaneously.

## Real-Life Analogy

A supermarket with many checkout counters can serve more customers.

## Interview Answer

Throughput is the number of requests or tokens processed per unit time.



# PART 2: LLM EVALUATION

Evaluation helps determine how good an LLM is at different tasks.



# Why Evaluation Is Needed

Evaluation helps answer:

* Which model is better?
* Which model is best at coding?
* Which model is best at reasoning?
* Which model performs better in conversations?


# 11. MMLU

## Full Form

Massive Multitask Language Understanding

## Purpose

Measures general knowledge and reasoning ability.

## Subjects Covered

* Mathematics
* Physics
* Chemistry
* Biology
* History
* Law
* Medicine
* Computer Science

## Example

Question:

What is the capital of France?

Answer:

Paris

## What It Measures

* General knowledge
* Reasoning
* Academic understanding

## Limitation

A high MMLU score does not guarantee strong coding or conversation ability.



MMLU evaluates an LLM's performance across multiple academic subjects.



# 12. GSM8K

## Full Form

Grade School Math 8K

## Purpose

Measures mathematical reasoning ability.

## Example

5 apples cost ₹50.

What is the cost of 10 apples?

Answer:

₹100

## What It Measures

* Arithmetic
* Logical reasoning
* Step-by-step problem solving



GSM8K evaluates mathematical reasoning using grade-school-level problems.



# 13. HumanEval

## Purpose

Measures coding ability.

## Example

Prompt:

Write a function that returns the square of a number.

Expected Code:

def square(x):
return x * x

## What It Measures

* Programming skills
* Algorithm design
* Code correctness
* Problem solving

## Why Important

Generated code is tested against hidden test cases.



HumanEval evaluates code generation and programming capability.



# 14. MT-Bench

## Full Form

Multi-Turn Benchmark

## Purpose

Measures conversational performance.

## Example

User:

Explain AI.

User:

Now explain it to a child.

The model must remember previous context.

## What It Measures

* Context retention
* Conversation quality
* Instruction following
* Multi-turn reasoning



MT-Bench evaluates how well a model performs in multi-turn conversations.


# 15. Arena Scores

## Definition

Arena Scores are rankings based on human preference.

## How It Works

User asks a question.

Two models generate answers.

Humans vote for the better answer.

Rankings are updated.

## What It Measures

* Helpfulness
* Accuracy
* Clarity
* User satisfaction

## Why Important

Human judgment often reflects real-world usefulness better than benchmark scores.


Arena Scores measure model quality based on human voting and preferences.
