API (Application Programming Interface)
Definition :
API is a interface or bridge that allows two software applications to communicate and exchange data with each other.

How API Works
API mainly works using Request and Response.
1. Request
A request is sent by the client (app/user) asking for some data or service.
Example: "Give me today's weather in Visakhapatnam."
2. Response
A response is the data returned by the server through the API.
Example: Temperature: 32°C
Humidity: 75%

 Example
* Restaurant Example
* Customer = User/App
* Waiter = API
* Kitchen = Server

Flow:
* Customer → Gives Order (Request)
* Waiter → Takes Order to Kitchen
* Kitchen → Prepares Food
* Waiter → Brings Food Back (Response)
* Customer → Receives Food

Uses of API
* Fetch data from servers.
* Connect different applications.
* Get real-time information (weather, maps, news).
* Enable login using Google or Facebook.
Process online payments.
