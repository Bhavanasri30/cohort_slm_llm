# Parameters

Think of them as the model's memory.

This knowledge is stored inside billions of parameters.

More parameters usually mean:

More knowledge
Better understanding
Higher cost

# Attention Mechanism

Attention allows the model to focus on important words.

Example:

Ravi went to school because he had an exam.

The model understands:

he = Ravi

This understanding comes from attention.





A Transformer is the core technology used in modern LLMs like GPT and Gemini.
It first converts words into numbers and then uses attention to find important relationships between words.
This helps the model understand the meaning and context of a sentence.
Based on this understanding, it predicts the next word and generates responses.
It is called the **heart of an LLM** because it performs the main understanding, reasoning, and text-generation tasks.













# SLM (Small Language Model)


A Small Language Model (SLM) is a lightweight AI model with fewer parameters,designed to perform language tasks efficiently using less computational resources.
* It is designed to understand and generate text using less memory and cost
* It is Benefitial for  Faster response time,Lower cost,Less memory usage,can run on laptops, phones, and edge devices,better for privacy-sensitive applications

 How slm works:
Input Text
     ↓
Transformer Architecture
     ↓
Parameters (Knowledge)
     ↓
Predict Next Word
     ↓
Generate Output

# Advantages of SLMs
* Speed:Generate responses quickly.

* Cost Efficiency:Cheaper to train and run.

* Privacy:Can run locally without sending data to external servers.

* Lower Hardware Requirements:Suitable for devices with limited resources.

* Easy Deployment:Simpler to deploy for small-scale applications.

# limitations of SLMs
* Less Knowledge:may know less than larger models.

* Lower Reasoning Ability:Complex reasoning tasks can be more challenging.

* Smaller Context Understanding:May struggle with very large documents or highly complex tasks.

# Embedding
Definition:

An embedding is a numerical representation of text that captures its meaning.

Example

Suppose we convert words into vectors:

Dog   → [0.8, 0.6, 0.2]
Puppy → [0.82, 0.58, 0.25]
Car   → [0.1, 0.9, 0.7]

Notice:

Dog ≈ Puppy

Their vectors are close together.

But:

Dog ≠ Car

Their vectors are far apart


#Vector Database

Definition

A Vector Database stores embeddings and quickly finds similar embeddings.

# Vector Database Example

Imagine you have 3 documents:

Doc 1: Dogs are loyal animals.
Doc 2: Cats like sleeping.
Doc 3: Cars need fuel.

After creating embeddings:

Doc 1 → [10, 20]
Doc 2 → [12, 18]
Doc 3 → [90, 95]

User asks:
Tell me about puppies.

Embedding Model converts it to:

Puppies → [11,19]

Vector Database compares distances:

Puppies → Doc 1 = Very Close ✅
Puppies → Doc 2 = Close
Puppies → Doc 3 = Very Far

So it returns:

Doc 1: Dogs are loyal animals.
