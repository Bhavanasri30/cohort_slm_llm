```python
text = "I love AI"

tokens = text.split()

token_ids = {token: idx for idx, token in enumerate(set(tokens), start=1)}

print("Tokens:", tokens)
print("Token IDs:", token_ids)
```

* text = "I love AI"

Purpose: Stores the input sentence that we want to tokenize.

* tokens = text.split()

Purpose: Splits the sentence into individual words (tokens) using spaces.

* token_ids = {token: idx for idx, token in enumerate(set(tokens), start=1)}

Purpose: Creates a dictionary and assigns a unique numerical ID to each token.

* set(tokens) → Removes duplicate words.
* enumerate(..., start=1) → Assigns numbers starting from 1.
* {token: idx} → Creates a Token → ID mapping.

Example:

I → 1
love → 2
AI → 3
* print("Tokens:", tokens)

Purpose: Displays the list of tokens generated from the sentence.

Output:

Tokens: ['I', 'love', 'AI']

* print("Token IDs:", token_ids)

Purpose: Displays the token-to-ID mapping.

Output:

Token IDs: {'I': 1, 'love': 2, 'AI': 3}
