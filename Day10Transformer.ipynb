# Transformer

Definition:

A Transformer is a model that reads all words in a sentence at the same time and learns how each word is related to every other word.

# Why Were Transformers Introduced?

Transformers were introduced to overcome the limitations of older sequence models like RNNs (Recurrent Neural Networks) and LSTMs (Long Short-Term Memory networks)

* Slow training

*  Could forget information from long sentences

* Difficult to process large amounts of text.



When reading a long sentence, older models could forget the meaning of earlier words.

Transformers solved this problem.


# Transformer Basic Architecture Overview

A Transformer processes text through several stages to understand meaning and generate output.

Architecture Flow

* Input Text
    
* Tokenizer
    
* Embeddings
    
* Positional Encoding
    
* Multi-Head Attention
    
* Feed Forward Network
    
* Output


**1. Input Text**

The user provides a sentence.

Example:

I love AI.


**2. Tokenizer**

Breaks the sentence into smaller pieces called tokens.

["I", "love", "AI"]


**3. Embeddings**

Converts each token into numerical vectors so the computer can process them.

* I    → [0.2, 0.5, 0.7]
* love → [0.8, 0.3, 0.4]
* AI   → [0.9, 0.6, 0.2]


**4. Positional Encoding**

Adds position information because Transformers process all words together.

* I    = Position 1
* love = Position 2
* AI   = Position 3

This helps the model know the order of words.


**5. Multi-Head Attention**

The most important part.

Each word looks at all other words and learns which ones are important.

Example:

The cat drank milk because it was hungry.

Attention mechanism helps the model understand:

it → cat



# Attention mechanism:


**6. Feed Forward Network**

Processes the information learned by attention and identifies patterns.

Think of it as refining and improving understanding.

**7. Output Layer**

Produces the final result.

Examples:

* Answer a question
* Translate text
* Summarize text
* Predict the next word


The basic architecture of a Transformer consists of Tokenization, Embeddings, Positional Encoding, Multi-Head Attention, Feed Forward Networks, and an Output Layer. The input text is first converted into tokens and embeddings. Positional encoding adds word-order information. Attention helps the model understand relationships between words. Finally, the feed-forward network processes this information and generates the output.
